```python
# instantiation 예제

class Dog:  
# 개를 표현하는 클래스

    def __init__(self, name, breed):
        # name breed 속성을 초기화
        self.name = name
        self.breed = breed
    def sit(self):
        # 앉는 동작을 출력
        print(f"{self.name} 앉아 있습니다!")
    
    def roll_over(self):
        # 구르는 동작을 출력
        print(f"{self.name} 구르고 있습니다!")
```

```python
my_dog = Dog('바둑이', '진돗개')  # Dog 클래스의 인스턴스 생성
print(f"내 개의 이름은 {my_dog.name}이고, 품종은 {my_dog.breed}입니다.")
my_dog.sit()          # sit 메서드 호출
my_dog.roll_over()    # roll_over 메서드 호출 
your_dog = Dog('토이', '푸들')
print(f"너의 개의 이름은 {your_dog.name}이고, 품종은 {your_dog.breed}입니다.")
your_dog.sit()
your_dog.roll_over()    # roll_over 메서드 호출 
# 개의 인스턴스 생성 및 메서드 호출 예제
# 개를 표현하는 클래스
```
