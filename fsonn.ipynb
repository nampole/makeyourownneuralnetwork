```python
# 객체
class Dog:
# 개는짖을수있습니다(bark). 
    def bark(self):
        print("woof!")
        pass
# 개는 앉을수있습니다(sit).
    def sit(self):
        print("sitting")
        pass
    pass
myDog = Dog()
yourDog = Dog()
myDog.bark()
myDog.sit() 
yourDog.bark()
yourDog.sit()   
```

```python
class Dog:
# initialisation method with internal data
    def __init__(self, petname, temp): 
        self.name = petname; 
        self.temperature = temp;
# get status
    def status(self):
        print("dog name is", self.name)
        print ("dog temperature", self.temperature)
# set temperature
    def setTemperature(self, temp):
        self.temperature = temp;
        pass
# dogs can bark() 
    def bark(self): 
        print ("멍멍!") 
        pass
pass
fidoDog = Dog("Fido", 38.5)
buddyDog = Dog("Buddy", 39.0)
fidoDog.status()
buddyDog.status()
fidoDog.setTemperature(37.8)
fidoDog.bark()
fidoDog.status()    
```
